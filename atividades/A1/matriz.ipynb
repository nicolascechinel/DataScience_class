```python
# Solicita ao usuário o nome do arquivo para salvar o conteúdo
nome_arquivo = input("Digite o nome do arquivo para salvar o conteúdo: ")

# Abre o arquivo no modo de escrita
with open(nome_arquivo, 'w') as arquivo:
    while True:
        # Solicita ao usuário o conteúdo a ser armazenado
        conteudo = input("Digite o conteúdo (ou 'sair' para encerrar): ")

        # Verifica se o usuário deseja sair
        if conteudo.lower() == 'sair':
            break

        # Escreve o conteúdo no arquivo
        arquivo.write(conteudo + '\n')

print("Conteúdo foi armazenado no arquivo com sucesso!")
```

```python
from collections import Counter

# Função para ler o arquivo e contar as palavras
def count_unique_words(filename):
    unique_word_counter = Counter()  # Usaremos o Counter para contar palavras únicas
    
    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()  # Separando palavras da linha
            unique_word_counter.update(words)  # Atualizando o contador com as palavras da linha
    
    return unique_word_counter

# Nome do arquivo
filename = 'ciencia_de_dados.txt'

# Contando palavras únicas
word_counter = count_unique_words(filename)

# Exibindo a contagem de palavras únicas
for word, count in word_counter.items():
    print(f'{word}: {count}')
```
