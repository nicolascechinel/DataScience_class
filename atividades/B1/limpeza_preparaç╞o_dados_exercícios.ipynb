```python
import pandas as pd
```

### 1) Carregue o arquivo books.csv utilizando o pandas.

```python
data = pd.DataFrame(pd.read_csv ('books.csv', sep = ','))
print(data)
```

### 2) Elimine as colunas 'Corporate Author', 'Corporate Contributors', 'Former owner' e 'Engraver'.

```python
data = data.drop(['Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver'], axis=1)
```

### 3) Verifique se a coluna 'Identifier' é única e, em caso afirmativo, atribua esta coluna como índice do conjunto de dados.

```python
print(data['Identifier'].is_unique)
```

### 4) Extrair a data de publicação de modo que somente o ano com quatro posições seja recuperado. Armazene o resultado em uma variável.


```python

resutado = data['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
print(resutado)
```

### 5) Inclua a nova coluna (armazenada no exercício anterior) no conjunto de dados e, após isso, elimine a coluna 'Date of Publication'.

```python
#Inclua a nova coluna (armazenada no exercício anterior) no conjunto de dados e, após isso, elimine a coluna 'Date of Publication'
data['New Date of Publication'] = resutado
data = data.drop(['Date of Publication'], axis=1)
print(data)
```

### 6) Salve o resultado do processamento em um arquivo csv.

```python
data.to_csv('books_limpo.csv', sep=',', index=False)
```

```python
print(data.groupby('New Date of Publication').count())
```
