```python
# Carrega as bibliotecas
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier # importa o Decision Tree Classifier
from sklearn.model_selection import train_test_split # importa o método train_test_split 
from sklearn import metrics #importa as métricas de avaliação de modelos do scikit-learn
from sklearn.model_selection import cross_validate # importa o método de cross validation
from sklearn import preprocessing
```

```python
# Define as colunas
col_names = ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']
# Carrega o dataset
df = pd.read_csv("drug200.csv", header=None, names=col_names)
#label unacc=2, acc=0, good=1, v-good=3

#cria uma instância de labelEncoder
le = preprocessing.LabelEncoder()
# Converte strings em números
df['Drug'] = le.fit_transform(df['Drug'].astype(str))
df['BP'] = le.fit_transform(df['BP'].astype(str))
df['Cholesterol'] = le.fit_transform(df['Cholesterol'].astype(str))
df['Sex'] = le.fit_transform(df['Sex'].astype(str))
df

```

```python
df.head()
```

```python
#Separa o conjunto de dados em atributos previsores e atributo alvo/meta
feature_cols = ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']
X = df[feature_cols] # Características (atributos previsores)
y = df.Drug # Variável alvo/meta (atributo alvo)
print(X)
print(y)
```

```python
# Separa o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 60% treinamento e 40% teste
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
# Importa o classificador K-Neighbors
from sklearn.neighbors import KNeighborsClassifier

#Cria uma instância do classificador K-NN
knn = KNeighborsClassifier(n_neighbors=5)

#Treina o modelo
knn.fit(X_train, y_train)

#Realiza a predição com o conjunto de teste
y_pred = knn.predict(X_test)
```

```python
# Apresenta a acurácia do modelo
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Importa o classificador GaussianNB
from sklearn.naive_bayes import GaussianNB

# Cria uma instância do classificador
gnb = GaussianNB()

# Treina o modelo 
gnb.fit(X_train, y_train)

#Realiza a predição com o conjunto de teste
y_pred = gnb.predict(X_test)
```

```python
# Apresenta a acurácia do modelo
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Importa o classificador Random Forest
from sklearn.ensemble import RandomForestClassifier

# Cria uma instância classificador
rfc=RandomForestClassifier(n_estimators=100)

# Treina o modelo usando os conjuntos de treinamento (características e rótulo para predição)
rfc.fit(X_train,y_train)

#Realiza a predição com o conjunto de teste
y_pred=rfc.predict(X_test)
```

```python
# Apresenta a acurácia do modelo
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
