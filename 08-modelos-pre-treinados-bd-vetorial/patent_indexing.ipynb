```python
!conda install pytorch -y
!conda install sentence-transformers -y
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from elasticsearch import Elasticsearch
model = SentenceTransformer('all-MiniLM-L6-v2') #Loads the pre-trained SBERT model  
from elasticsearch.helpers import bulk
```

```python
elastic_client = Elasticsearch(hosts=['http://localhost:9200'], basic_auth=('elastic', 'master'))
```

### 1_dataset_title_abstract_600_16_treinamento

```python
patent_data = pd.read_csv("datasets/1_dataset_title_abstract_600_16_treinamento.csv")
rows = len(patent_data.index)
index_name  = "patents_1"
print(patent_data.head(5))
print(rows)
print(index_name)
```

```python
def build_sbert_doc(text:str, examiner:str):
  document={ 
    "examiner": examiner,
    "text": text,
    "embedding": model.encode(text).tolist()
  }   
  return document
```

```python
ctr = 0
actions = []
for index, row in patent_data.iterrows():
    ctr = ctr + 1
    actions.append({"_index":index_name.lower(), 
                    "_source": build_sbert_doc(row.text, row.examiner)})
    if ((ctr % 100 ) == 0):
      bulk(elastic_client, actions)
      actions.clear()
      print("{} documents were indexed!!!".format(ctr))

if len(actions) != 0:
  bulk(elastic_client, actions)
  actions.clear()
  print("{} documents were indexed!!!".format(ctr))
print("Indexing finished successfully!!!")
```
